```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "beer_reviews.csv"
beer_reviews_df = pd.read_csv(csv_file)
beer_reviews_df.head()
```

```python
beer_reviews_df.count()
```

```python
beer_reviews_clean = beer_reviews_df.dropna()
```

```python
beer_reviews_clean.count()
```

```python
beer_reviews_clean.head()
```

```python
beer_reviews_clean2 = beer_reviews_clean.drop(columns=['brewery_id', 'review_time','review_profilename', 'beer_beerid' ])
```

```python
beer_reviews_clean2.head()
```

```python
beer_reviews_clean3 = beer_reviews_clean2.rename(columns={"brewery_name": "Brewery",
                                                          "review_overall": "Overall Review",
                                                         "review_aroma": "Aroma Review",
                                                         "review_appearance": "Appearance Review",
                                                         "beer_style": "Beer Style",
                                                          "review_palate": "Palate Review",
                                                          "review_taste": "Taste Review",
                                                          "beer_name": "Beer Name",
                                                          "beer_abv": "Alcohol By Volume"})
```

```python
beer_reviews_clean3.head()
```

```python
beer_reviews_clean3.to_csv(r'beer_reviews_final.csv', index = False )
```
