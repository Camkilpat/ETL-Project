```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
breweries = "beer_reviews_final.csv"
location = "Location.csv"


breweries_df = pd.read_csv(breweries)
location_df = pd.read_csv(location)

```

```python
breweries_df11 = breweries_df.groupby(['Brewery','Beer Name']).mean()[['Overall Review','Aroma Review','Appearance Review', 'Palate Review','Taste Review']]
breweries_df12 = pd.DataFrame (breweries_df11)
breweries_df13 = pd.merge(breweries_df12, breweries_df, on = (['Brewery','Beer Name']), how = 'left')
breweries_df13
```

```python
breweries_df14 = breweries_df13.drop_duplicates(subset = ['Beer Name'])
breweries_df14.count()
```

```python
breweries_df15 = breweries_df14.drop(columns=['Overall Review_y','Aroma Review_y', 'Appearance Review_y','Palate Review_y', 'Taste Review_y'])
breweries_df15.head()
```

```python
breweries_df16 = breweries_df15.rename(columns={'Overall Review_x':'Overall Review','Aroma Review_x':'Aroma Review', 'Appearance Review_x':'Appearance Review','Palate Review_x':'Palate Review', 'Taste Review_x':'Taste Review'})
breweries_df16
```

```python
breweries_df17 = pd.merge(breweries_df16, location_df,
                                 how='left', on='Brewery')
breweries_df17
```

```python
breweries_df18 = breweries_df17.dropna()
breweries_df18
```

```python
breweries_df19 = breweries_df18.round()
breweries_df19
```

```python
breweries_df19.to_csv('breweries_final.csv', index= False)
```
